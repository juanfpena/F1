```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib import cm
import numpy as np
import pandas as pd

ff1.Cache.enable_cache("cache")
plotting.setup_mpl()
```

```python
qa = ff1.get_session(2022, "Bahrain", "Q")
qa.load(telemetry=True)
```

```python
qp = ff1.get_session(2022, "Bahrain", "R")
qp.load(telemetry=True)
```

```python
actual_lec = qp.laps.pick_driver("SAI").pick_fastest().get_car_data().add_distance()
actual_ver = qp.laps.pick_driver("PER").pick_fastest().get_car_data().add_distance()
```

```python
actual_lec["Session"] = "SAI"
actual_ver["Session"] = "PER"

leclerc = pd.concat([actual_lec, actual_ver], axis=0)
```

```python
import plotly.express as px

colors = {"SAI": "#dc0000", "PER": "#0600ef"}

fig = px.line(leclerc, "Distance", "Speed", color="Session", color_discrete_map=colors)
fig.show()
```

```python
fig = px.line(leclerc, "Distance", "Throttle", color="Session")
fig.show()
```

```python
fig = px.line(leclerc, "Distance", "nGear", color="Session")
fig.show()
```
