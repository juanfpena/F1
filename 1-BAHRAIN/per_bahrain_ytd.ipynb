```python
import fastf1 as ff1
from fastf1 import plotting
import pandas as pd

ff1.Cache.enable_cache("cache")
plotting.setup_mpl()
```

```python
qa = ff1.get_session(2022, "Bahrain", "R")
qa.load(telemetry=True)
```

```python
qp = ff1.get_session(2021, "Bahrain", "R")
qp.load(telemetry=True)
```

```python
actual_perez = qa.laps.pick_driver("PER").pick_fastest().get_car_data().add_distance()
past_perez = qp.laps.pick_driver("PER").pick_fastest().get_car_data().add_distance()
```

```python
actual_perez
```

```python
actual_perez["Session"] = "2022"
past_perez["Session"] = "2021"

perez = pd.concat([actual_perez, past_perez], axis=0)
```

```python
import plotly.express as px

fig = px.line(perez, "Distance", "Speed", color="Session")
fig.show()
```

```python
fig = px.line(perez, "Distance", "Throttle", color="Session")
fig.show()
```

```python
fig = px.line(perez, "Distance", "nGear", color="Session")
fig.show()
```
