```python
import pandas as pd

pd.set_option("display.max_rows", 200)

constructor_results = pd.read_csv("archive/constructor_results.csv")
constructors = pd.read_csv("archive/constructors.csv")
races = pd.read_csv("archive/races.csv")
results = pd.read_csv("archive/results.csv")
```

```python
c = constructor_results.merge(constructors, on="constructorId")
c = c[["constructorResultsId", "raceId", "points", "constructorRef", "name"]]
```

```python
current_constructors = ['Alfa Romeo','AlphaTauri','Alpine F1 Team','Aston Martin','Ferrari','Haas F1 Team','McLaren','Mercedes','Red Bull','Williams']

c = c[c.name.isin(current_constructors)]
```
