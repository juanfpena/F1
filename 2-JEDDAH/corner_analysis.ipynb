```python
import numpy as np
import pandas as pd

import fastf1 as ff1
from fastf1 import plotting

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
ff1.Cache.enable_cache("cache")

test = ff1.get_session(2022, "Jeddah", 3)
test.load(telemetry=True)
```

```python
fastest_perez = test.laps.pick_driver("PER").pick_fastest().get_car_data().add_distance()
fastest_sainz = test.laps.pick_driver("SAI").pick_fastest().get_car_data().add_distance()

fastest_sainz["DRIVER"] = "Sainz"
fastest_perez["DRIVER"] = "Perez"

speed_analysis = pd.concat([fastest_perez, fastest_sainz], axis=0)[["Speed", "Distance", "Throttle", "DRIVER"]]
```

```python
fastest = speed_analysis[(speed_analysis.Distance < 750) & (speed_analysis.Distance > 250)]
slowest = speed_analysis[(speed_analysis.Distance < 2000) & (speed_analysis.Distance > 1500)]
```
